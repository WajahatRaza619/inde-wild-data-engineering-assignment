```python
import pandas as pd

df = pd.read_csv("01-31-Jan-2026-Zepto.csv")

print(df.head())
```

```python
df.head(10)
```

```python
df_zepto=df.copy()
```

```python
df_zepto['date']=pd.to_datetime(df_zepto['Date'], dayfirst=True)
```

```python
df_zepto['product_identifier'] = df_zepto['SKU Number']
df_zepto['total_units'] = df_zepto['Sales (Qty) - Units']
df_zepto['total_revenue'] = df_zepto['Gross Merchandise Value']
```

```python
df_zepto['data_source'] = 'Zepto'
```

```python
df_zepto = df_zepto[['date', 'product_identifier', 'total_units', 'total_revenue', 'data_source']]
```

```python
df_zepto.head(10)
```

```python
print(df_zepto.isnull().sum())
```

```python
print(df_zepto.dtypes)
```

```python
df_myntra = pd.read_csv("myntra-jan26.csv")

print(df_myntra.head())
print(df_myntra.columns)
print(df_myntra.info())
```

```python
df_myntra_clean = df_myntra.copy()

# Convert date
df_myntra_clean['date'] = pd.to_datetime(
    df_myntra_clean['order_created_date'],
    format='%Y%m%d',
    errors='coerce'
)

# Standardize columns
df_myntra_clean['product_identifier'] = df_myntra_clean['style_id']

df_myntra_clean['total_units'] = df_myntra_clean['sales']

df_myntra_clean['total_revenue'] = df_myntra_clean['mrp_revenue']

# Add source
df_myntra_clean['data_source'] = 'Myntra'

# Select required columns
df_myntra_clean = df_myntra_clean[
    ['date', 'product_identifier', 'total_units', 'total_revenue', 'data_source']
]
```

```python
print(df_myntra_clean.head())
print(df_myntra_clean.isnull().sum())
print(df_myntra_clean.dtypes)
```

```python
df_nykaa = pd.read_csv("1-31-Jan-Nykaa-online.csv")

print(df_nykaa.head())
print(df_nykaa.columns)
print(df_nykaa.info())
```

```python
df_nykaa_clean = df_nykaa.copy()

# Convert date
df_nykaa_clean['date'] = pd.to_datetime(df_nykaa_clean['date'], errors='coerce')

# Standardize columns
df_nykaa_clean['product_identifier'] = df_nykaa_clean['SKU Code']

df_nykaa_clean['total_units'] = df_nykaa_clean['Total Qty']

# 🔥 Important: Calculate revenue
df_nykaa_clean['total_revenue'] = (
    df_nykaa_clean['Selling Price'] * df_nykaa_clean['Total Qty']
)

# Add source
df_nykaa_clean['data_source'] = 'Nykaa'

# Select required columns
df_nykaa_clean = df_nykaa_clean[
    ['date', 'product_identifier', 'total_units', 'total_revenue', 'data_source']
]
```

```python
print(df_nykaa_clean.head())
print(df_nykaa_clean.isnull().sum())
print(df_nykaa_clean.dtypes)
```

```python
df_blinkit = pd.read_csv("01-31-Jan-2026-Blinkit-Sales.csv")

print(df_blinkit.head())
print(df_blinkit.columns)
print(df_blinkit.info())
```

```python
df_blinkit_clean = df_blinkit.copy()

# Convert date
df_blinkit_clean['date'] = pd.to_datetime(df_blinkit_clean['date'], errors='coerce')

# Standardize columns
df_blinkit_clean['product_identifier'] = df_blinkit_clean['item_id']

df_blinkit_clean['total_units'] = df_blinkit_clean['qty_sold']

df_blinkit_clean['total_revenue'] = df_blinkit_clean['mrp']

# Add source
df_blinkit_clean['data_source'] = 'Blinkit'

# Select required columns
df_blinkit_clean = df_blinkit_clean[
    ['date', 'product_identifier', 'total_units', 'total_revenue', 'data_source']
]
```

```python
print(df_blinkit_clean.head())
print(df_blinkit_clean.isnull().sum())
print(df_blinkit_clean.dtypes)
```

```python
final_df = pd.concat([
    df_zepto,
    df_myntra_clean,
    df_nykaa_clean,
    df_blinkit_clean
])
```

```python
final_df = final_df.groupby(
    ['date', 'product_identifier', 'data_source']
).agg({
    'total_units': 'sum',
    'total_revenue': 'sum'
}).reset_index()
```

```python
final_df.rename(columns={
    'product_identifier': 'product_identifier(sku)'
}, inplace=True)
```

```python
final_df = final_df[
    ['date', 'product_identifier(sku)', 'total_units', 'total_revenue', 'data_source']
]
```

```python
final_df.to_csv("final_sales.csv", index=False)
```
